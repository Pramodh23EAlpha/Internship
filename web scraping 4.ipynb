##  Q1
Scrape the details of most viewed videos on YouTube from Wikipedia.

A) Rank

C) Artist
D) Upload date

```python
# Activating the chrome browser
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(3)

YtRanks = []
YtNames = []
YtArtist = []
YtUploadDate = []
YtViews = []


scrapdata_ranks = driver.find_elements_by_xpath('//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
scrapdata_name = driver.find_elements_by_xpath('//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
scrapdata_uploaader = driver.find_elements_by_xpath('//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
scrapdata_views = driver.find_elements_by_xpath('//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
scrapdata_uploaddate = driver.find_elements_by_xpath('//table[@class = "wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')

for val  in scrapdata_ranks:
    YtRanks.append(val.text)
    
for val  in scrapdata_name:
    YtNames.append(val.text)
    
for val  in scrapdata_uploaader:
    YtArtist.append(val.text)

for val  in scrapdata_views:
    YtViews.append(val.text)    
    
for val  in scrapdata_uploaddate:
    YtUploadDate.append(val.text) 
    
ytdf =pd.DataFrame({})
ytdf['Ranks']=YtRanks
ytdf['Name']=YtNames
ytdf['Artist']=YtArtist
ytdf['UploadDates']=YtUploadDate
ytdf['Views']=YtViews
ytdf    
```

## Q5. 
1Scrape the details of trending repositories on Github.com.

Url = https://github.com/

You have to find the following details:

A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
# Activating the chrome browser
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
time.sleep(3)

titles = []
description = []
contributorscount = []
languages_list = []

# Opening the homepage of 
url = "https://github.com/"
driver.get(url)
time.sleep(3)
driver.get(str(driver.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a').get_attribute('href')))
time.sleep(3)

urls = driver.find_elements_by_xpath('//h1[@class = "h3 lh-condensed"]/a')

url_list = []
for url in urls:
    url_list.append(url.get_attribute('href'))
    
    
for data in url_list: 
    driver.get(data)
    time.sleep(3)
    titles.append(driver.find_element_by_xpath('//*[@id="repository-container-header"]/div[1]/div/h1/strong/a').text)
    try:
        description.append(driver.find_element_by_xpath('//p[@class = "f4 my-3"]').text)
    except NoSuchElementException:
        description.append('-')       
    contributorscount.append(driver.find_element_by_xpath('//div/h2/a[@class = "Link--primary no-underline"]').text)
    languages = driver.find_elements_by_xpath('//div[@class = "f6"]/a')
    language = ""
    for lang in languages:
        language = language + lang.text + "|"
        #print(language)
    languages_list.append(language)
    
languages_list  

gitdf =pd.DataFrame({})
gitdf['Title']=titles
gitdf['Description']=description
gitdf['Contributors count']=contributorscount
gitdf['Languages']=languages_list


gitdf
```

## 10.

```python
# Activating the chrome browser
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
# Opening the homepage of 
url = "https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(3)

driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]').click()
time.sleep(3)

uci_names = []
uci_datatypes = []
uci_task = []
uci_attributetype = []
uci_noofinstances = []
uci_noofattribute = []
uci_year = []

count = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[1]/tbody/tr/td[1]/p/b').text

print(int(count))

for i in range(int(count)):
    
    try:

        uci_names.append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i+2)+']/td[1]/table/tbody/tr/td[2]/p/b/a').text)
        uci_datatypes.append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i+2)+']/td[2]/p').text)
        uci_task.append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i+2)+']/td[3]/p').text)
        uci_attributetype.append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i+2)+']/td[4]/p').text)
        uci_noofinstances.append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i+2)+']/td[5]/p').text)
        uci_noofattribute.append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i+2)+']/td[6]/p').text)
        uci_year.append(driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr['+str(i+2)+']/td[7]/p').text)
        
    except:
        print('EXCEPTION')

        
ucidf =pd.DataFrame({})
ucidf[' Name ']=uci_names
ucidf['Data Types']=uci_datatypes
ucidf['Task ']=uci_task
ucidf['Attribute type']=uci_attributetype
ucidf['No of instances ']=uci_noofinstances
ucidf['No of attribute ']=uci_noofattribute
ucidf['Year']=uci_year

ucidf 
```
